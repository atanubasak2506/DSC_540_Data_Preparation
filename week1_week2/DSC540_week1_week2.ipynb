```python
#2. Create a Jupyter notebook where you create a list, iterate over the list and sort your results,
#generate random numbers, add to the list, and then print your results.

import random
lst = []
for i in range(0,10):
    lst.append(random.randint(0,50))
print(lst)

#Function that iterates over the list
def _iterate_list(lst):
    for i in lst:
        print(i)

#Function that sort the list
def _sort_list(lst):
    lst.sort()
    print(lst)
    
print('iterating list'+'='*50)
_iterate_list(lst)

print('sorting list'+'='*50)
_sort_list(lst)
```

```python

#a. Data Wrangling with Python: Activity 1 page 17

import random
list1 = random.sample(range(1, 500), 100)
print("list1 : "+str(list1))

list2 = [x for x in list1 if x % 3==0]
print("list2 : "+str(list2))

diff_list_count = len(list1) - len(list2) 

print("diff_list_count : {}".format(diff_list_count))
```

```python
#perform step 2 and 3 using a loop

diff_list = []
for x in range(3):
    
    print('=================> loop '+str(x+1))
    list1 = random.sample(range(1, 500), 100)
    list2 = [x for x in list1 if x % 3==0]
    print("list2 : "+str(list2))

    diff_list_count = len(list1) - len(list2) 
    diff_list.append(diff_list_count)
    print("diff_list_count : {}".format(diff_list_count))
```

```python
from statistics import mean
print("Arithmatic mean of differences : "+str(mean(diff_list)))
```

```python
#b. Data Wrangling with Python: Activity 2 page 31
text = """Data science is growing up fast. Over the past five years companies have invested billions to get 
the most-talented data scientists to set up shop, amass zettabytes of material, and run it through their 
deduction machines to find signals in the unfathomable volume of noise. It’s working—to a point. Data has 
begun to change our relationship to fields as varied as language translation, retail, health care, and basketball."""
print(text)
#removing newlines
text1 = text.replace('\n','')
print(text1)
```

```python
#getting unique word and their count.
punc = "!@#$%^&*()[]{};:,./<>?\|`~-=_+"
for k in text1:
    if k in punc:
        text1=text1.replace(k,"")
word_list = text1.split()
print(word_list)
```

```python
word_count = {}
for k in word_list:
    if k not in word_count.keys():
        word_count[k] = 1
    else:
        word_count[k]+=1

print(word_count)
```

```python
word_count_wo_case = {}
for k in word_list:
    if k.lower() not in word_count_wo_case.keys():
        word_count_wo_case[k.lower()] = 1
    else:
        word_count_wo_case[k.lower()]+=1

print(word_count_wo_case)
```

```python
def permutations_to_number(lst):
    str_ = ''
    for i in lst:
        str_ = str_+str(i)
    return int(str_)
    
    
    
from itertools import dropwhile
num = [0,1,2]
p = permutations(num)

for j in list(p):
    str_to_int = permutations_to_number(list(dropwhile(lambda x : x == 0, list(j))))
    print(str_to_int)
```

```python
#d. Data Wrangling with Python: Activity 4 page 59

from itertools import zip_longest
def return_dict_from_csv_line(header, line):
    # Zip them
    zipped_line = zip_longest(header, line, fillvalue=None)
    # Use dict comprehension to generate the final dict
    ret_dict = {kv[0]: kv[1] for kv in zipped_line}
    return ret_dict
```

```python
with open("sales_record.csv", "r") as fd:
    first_line = fd.readline()
    print('='*100)
    header = first_line.replace("\n", "").split(",")
    #this is the header list from the csv
    print(header)
    print('='*100)
    for i, line in enumerate(fd):
        line = line.replace("\n", "").split(",")
        d = return_dict_from_csv_line(header, line)
        print(d)
        if i > 5:
            break
```
