# Format data into a more readable format

```python
df[['CAFV_Eligibility']].value_counts()
```

```python
def CAFV_Eligibility(x):
    if x=='Eligible':
        return 'Y'
    elif x=='Not Eligible':
        return 'N'
    else:
        return 'X'
    
df['CAFV_Eligibility'] = df['CAFV_Eligibility'].apply(CAFV_Eligibility)
```
