# Step 3 Update the index to a more readable format

```python
col_df['index'].hist()
```

```python
print(col_df['index'].min(), col_df['index'].max() )
```

```python
def get_index(row):
    if row < 115:
        return 'low'
    elif row < 150:
        return 'medium'
    else:
        return 'high'
    
col_df['index_level'] = col_df['index'].apply(get_index)
col_df.head()
```

# Step 4 Update the transportation Index to a more readable format

```python
col_df['transportation'].hist()
```

```python
def get_trans(row):
    if row < 95:
        return 'low'
    elif row < 100:
        return 'medium'
    else:
        return 'high'
    
col_df['transportation_level'] = col_df['transportation'].apply(get_trans)
col_df.head()
```

# Step 5 Update the housing index to a more readable format

```python
col_df['housing'].hist()
```

```python
def get_housing(row):
    if row < 95:
        return 'low'
    elif row < 100:
        return 'medium'
    else:
        return 'high'
    
col_df['housing_level'] = col_df['housing'].apply(get_housing)
col_df.head()
```
