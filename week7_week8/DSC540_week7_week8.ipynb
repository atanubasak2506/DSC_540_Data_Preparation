# Chapter 10 : Grouping with Dicts/Series

```python
candy_=candy.dropna(subset=['Q2: GENDER', 'Q3: AGE'])
```

```python
candy_ = candy_[candy_['Q3: AGE'].apply(lambda x: x.isnumeric())]
```

```python
candy_[['Q3: AGE']] = candy_[['Q3: AGE']].astype(int)
```

```python
candy_agg = candy_.groupby('Q2: GENDER').agg({'Q3: AGE' : 'mean'})
candy_agg.head()
```

# Chapter 10 : Grouping with Functions

```python
def max_age(x):
    return x.max()

candy_max_age_by_gender = candy_.groupby('Q2: GENDER').agg({'Q3: AGE' : max_age})
candy_max_age_by_gender.head()
```

# Chapter 11 : Convert between string and date time

```python
import pandas as pd
bb = pd.read_excel("BOING-BOING-CANDY-HIERARCHY-2016-SURVEY-Responses.xlsx")
bb.head(2)
```

```python
from datetime import datetime
def date_convert(date_to_convert):
     return datetime.strftime(date_to_convert,'%m/%d/%Y')
bb['ts_string'] = bb['Timestamp'].apply(date_convert)
bb.loc[:,['Timestamp','ts_string']].head()
```

```python
bb.loc[:,['Timestamp','ts_string']].dtypes
```
